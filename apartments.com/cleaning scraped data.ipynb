```python
import pandas as pd
import numpy as np
import re
data = pd.read_csv('LosAngeles.csv', error_bad_lines=False)
data.head()
```

```python
data.dtypes
prop=data["Property Info"]
```

```python
cleaned=[]
year=[]
units=[]
for i,line in enumerate (prop):
    temp = re.sub(r"[^0-9],/,"," ",str(line) ) 
    pop = re.findall(r"[0-9]{4}",temp)
    pop=re.sub(r"[^0-9],|\.|\(|\)|\[|\]|\'|\!|\@|\$|\%|\&|\*","",str(pop))
    dat=re.findall(r"(\d+)\ s*Units",temp)
    dat=re.sub(r"[^0-9],|\.|\(|\)|\[|\]|\'|\!|\@|\$|\%|\&|\*","",str(dat))
    cleaned.append(temp)
    year.append(pop)

    units.append(dat)
```

```python

                                               
data[' clean_property_info'] = cleaned
data['year']=year
data['units']=units


```
